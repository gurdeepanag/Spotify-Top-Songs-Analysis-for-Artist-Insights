# Data_Cleaning

## Importing_the_Data
Through the process of trial and error we discovered that our data could by decoded by "latin," and therefore we imported it accordingly.

We also only chose to import columns that would be necessary in answering our guiding questions, in order to save time and increase efficiency.

```python
path = "spotify-2023.csv"
dataset = pd.read_csv(path, encoding = "latin", usecols = ["track_name", "artist(s)_name", "artist_count", "in_spotify_playlists", "streams", "bpm", "key", "mode", "danceability_%", "valence_%", "energy_%", "acousticness_%", "instrumentalness_%", "liveness_%", "speechiness_%"])
display(dataset.head())
```

## Removing_Duplicate_Values

The method we chose to use in order to remove duplicate values, was to first order the dataset by number of streams in descending order, then check for rows with the same track name and artist name and keep only the duplicate row with the highest number of streams (which would be the first occurrence of the song).

We chose this methodology, as we believe that the song with the highest number of streams is the most up to date version on our dataset. We could not have merged the duplicate values because there are categorical variables could not be combined.

```python
dataset.sort_values(by = "streams", ascending = False, inplace = True)
dataset.drop_duplicates(subset = ["track_name", "artist(s)_name"], keep = "first", inplace = True)
# dropping "track_name" and "artist(s)_name" columns as we will not need them for the rest of our analyses
dataset.drop(columns = ["track_name", "artist(s)_name"], inplace = True)
```

## Removing_NaN_Values

We chose to remove NaN values by first checking the proportion of each column which contains NaN values, and if this was approximately 10% or greater, we would remove this column, as opposed to removing 10% or more of our dataset. We did this in order to preserve as much data as possible. In our case, the only column that had ~10% missing values, was the "key" feature, therefore we dropped it.

```python
# storing count of NaN values in each column
is_na = dataset.isna().sum()
# % of NaN values in each column
print(np.round(is_na[is_na > 0]/len(dataset), 2))
```

```python
dataset.drop(columns = ["key"], inplace = True)
```

## Renaming_Columns

We opted to rename our columns to more accurate and straight-to-the-point names, in order to make our code more readable. We chose to rename the "mode" column to "scale," as to avoid triggering the mathematical function "mode" when using dot notation.

```python
new_col_names = {"danceability_%": "danceability",
                "valence_%": "valence",
                "energy_%": "energy",
                "acousticness_%": "acousticness",
                "instrumentalness_%": "instrumentalness",
                "liveness_%": "liveness",
                "speechiness_%": "speechiness",
                "mode": "scale",
                "in_spotify_playlists": "num_playlists",
                "artist_count": "num_artists"
}
dataset.rename(columns = new_col_names, inplace = True)
```

## Checking_and_Changing_Column_Datatypes

We checked each of our column's datatypes in order to ensure our dataset is ready for numerical calculations and visualizations. First we noticed that the "streams" column was an "object" datatype, meaning that there must be a string datatype somewhere in it. The rest of the columns were the correct datatype for analyses.

```python
print(dataset.dtypes)
```

Next, we then decided to go through the "streams" column and check if each value is a numeric datatype if we apply an int() function to it, if it is not, then we would change it to a NaN value. Then we would drop any rows where the "streams" column is a NaN value.

```python
dataset.streams = dataset.streams.apply(lambda x: int(x) if x.isdigit() else np.nan)
dataset.dropna(subset = ["streams"], inplace=True)
print(dataset.dtypes)
```

Next, we decided to check the maximum value of each numerical feature in order to decide which datatype size would be appropriate for each column.

```python
max_columns = dataset.select_dtypes(include = [np.float64, np.int64]).max()
max_columns = max_columns.astype(np.int64)
display(max_columns)
```

We changed the datatypes of each column into smaller ones in order to reduce file memory size and increase efficiency.

```python
# maximum numeric value for int8: 127
# maximum numeric value for int16: 32,767
# maximum numeric value for int32: 2,147,483,647
# maximum numeric value for int64: (2^63) - 1

dataset.num_artists = dataset.num_artists.astype("int8")
dataset.num_playlists = dataset.num_playlists.astype("int32")
dataset.streams = dataset.streams.astype("int64")
dataset.bpm = dataset.bpm.astype("int16")
dataset.danceability = dataset.danceability.astype("int8")
dataset.valence = dataset.valence.astype("int8")
dataset.energy = dataset.energy.astype("int8")
dataset.acousticness = dataset.acousticness.astype("int8")
dataset.instrumentalness = dataset.instrumentalness.astype("int8")
dataset.liveness = dataset.liveness.astype("int8")
dataset.speechiness = dataset.speechiness.astype("int8")

print(dataset.dtypes)
```

## Changing_BPM_column_into_a_Categorical_Variable

In pursuit of answering our first guiding question, we would need to turn BPM into a categorical variable so that we may visualize its proportions later. To do so, we've begun by visualizing a histogram of the bpm column values to decide best on how to approach this. 

```python
fig = plt.figure(figsize=(10, 6))
fig.patch.set_facecolor("#191414")
plt.grid(visible = True)
plt.gca().set_axisbelow(True)
plt.hist(x=dataset["bpm"], bins=20, color="#1DB954", edgecolor='black')
plt.xticks(color='white')
plt.yticks(color='white')
plt.xlabel('Beats per Minute', color='white')
plt.ylabel('Frequency', color='white')
plt.title("BPM Histogram", color='white')
plt.tick_params(axis='x', colors='white')
plt.tick_params(axis='y', colors='white')
plt.show()
```

Based on the spread of values observed in this histogram, we decided on creating 3 bins for the BPM category column. We determined the bin width by taking the closest higher multiple of 10 to represent the maximum value and the closest lower multiple of 10 for the minimum value, subtracting these and floor dividing by 3. Then we created a list which went from the new lowest value to the new highest value with the step size of the range function being our newly calculated bin size. Essentially, the bin variable contains a list of four values (60, 110, 160 and 210), with the first and last value representing the minimum and maximum, and each interval representing a categorical variable.

```python
steps = 3
step_size = (np.max(dataset.bpm) // 10 * 10 + 10 - np.min(dataset.bpm) // 10 * 10) // steps
bins = list(range(np.min(dataset.bpm) // 10 * 10, np.max(dataset.bpm)// 10 * 10 + 10 + 1, step_size))
```

We re-visualized our BPM distribution, to illustrate our first, second and third intervals for our categorical variable naming convention (Low, Medium, High).

```python
fig = plt.figure(figsize=(10, 6))
fig.patch.set_facecolor("#191414")
plt.grid(visible = True)
plt.gca().set_axisbelow(True)
plt.hist(x=dataset["bpm"], bins=20, color="#1DB954", edgecolor='black')
plt.xticks(color='white')
plt.yticks(color='white')
plt.xlabel('Beats per Minute', color='white')
plt.ylabel('Frequency', color='white')
plt.vlines(x=bins, ymin=0, ymax=100, color="purple")
plt.title("BPM Histogram", color='white')
plt.tick_params(axis='x', colors='white')
plt.tick_params(axis='y', colors='white')
plt.show()
```

Here we created a new column, which labels each song's BPM category as being either Low, Medium or High based on our defined bins.

```python
labels = ['Low', 'Medium', 'High']
index_bpm = dataset.columns.get_loc("bpm")
bpm_category = pd.cut(dataset['bpm'], bins=bins, labels=labels)
dataset.insert(index_bpm + 1, "bpm_category", bpm_category)
display(dataset.head())
```

In pursuit of answering our second guiding question, we wanted to develop a better understanding of the range of values that existed in the number of streams feature. To do so, we've begun by visualizing a boxplot of the number of streams column values.


```python
# First we create the figure
fig = plt.figure(figsize=(10,6))
fig.patch.set_facecolor("#191414")
plt.grid(visible = True)
plt.gca().set_axisbelow(True)
# Create a box plot with a purple median line and green box
medianprops = {'color': 'purple'}
boxplot = plt.boxplot(x=dataset.streams, vert=False, medianprops=medianprops, patch_artist=True)

# Change the color of the box to green
for box in boxplot['boxes']:
    box.set(color='black', facecolor='#1DB954')

# Change the color of the outliers to purple
for flier in boxplot['fliers']:
    flier.set(marker='o', markerfacecolor='purple', markersize=6)

# Change the x and y label to white
plt.title("Boxplot of Number of Streams (Billions)")
plt.xlabel("Number of streams", color='white')
plt.ylabel("")
plt.grid(visible = True)
plt.gca().set_axisbelow(True)
plt.yticks(color='white')
plt.title("Number of Streams Boxplot", color='white')
plt.tick_params(axis='x', colors='white')
plt.tick_params(axis='y', colors='white')
plt.show()
```

To our surprise, we found that there were a significant number of outliers that extended past our upper whisker. These data points had significantly higher streams in contrast to the rest of the popular songs that it was sharing its "top-streaming" status with. These outliers introduced a strong right-skew to the distribution for the "number of streams" column and we felt as though their existence was telling us was that there was something unique about these songs in particular. This observation inspired us to slightly evolve and reframe our guiding questions to investigate these outliers which we decided to call "hyper-popular" in comparison to the rest of the songs which we labelled "popular" to represent songs that fell inside of the boxplot whiskers. To represent which songs were hyper-popular or not, we created a new column of boolean values that were True for songs that were hyper-popular (songs that fell beyond the upper whisker (Q3 + IQR) and False if not.

```python
# creating the hyper-popular cutoff value as the upper whisker of the boxplot
q1 = np.percentile(dataset["streams"], 25)
q3 = np.percentile(dataset["streams"], 75)
cutoff_hyperpopular = q3 + 1.5*(q3-q1)
# creating a new columns with boolean values, indicating whether a song is "hyper-popular" or not
dataset["is_hyper_popular"] = dataset["streams"].apply(lambda x: x > cutoff_hyperpopular)
```

```python
fig, ax = plt.subplots(nrows=7, ncols=2, figsize=(12, 30))
fig.patch.set_facecolor("#191414")

spotify_metrics = ["danceability", "valence", "energy", "acousticness", "instrumentalness", "liveness", "speechiness"]


for i, metric in enumerate(spotify_metrics):
    # Scatter plots for popular songs (hyper-popular status is False)
    ax[i][0].scatter(x=metric, y="streams", s=6, data=dataset[dataset.is_hyper_popular == False], color="#1DB954")
    b, a = np.polyfit(dataset[dataset.is_hyper_popular == False][metric], dataset[dataset.is_hyper_popular == False].streams, deg=1)
    ax[i][0].plot(dataset[dataset.is_hyper_popular == False][metric], a + b * dataset[dataset.is_hyper_popular == False][metric], color="purple")
    ax[i][0].set_xticks(range(0, 101, 10))
    ax[i][0].set_title(f"{metric.title()} vs Number of Streams for Popular Songs", color="white")
    ax[i][0].set_ylabel('Number of Streams (Billions)', color='white')
    ax[i][0].set_xlabel(f'{metric.title()} (%)', color='white')
    ax[i][0].grid(visible = True)
    ax[i][0].set_axisbelow(True)
    ax[i][0].tick_params(axis='x', colors='white')
    ax[i][0].tick_params(axis='y', colors='white')
    

    # Scatter plots for hyper-popular songs (hyper-popular status is True)
    ax[i][1].scatter(x=metric, y="streams", s=6, data=dataset[dataset.is_hyper_popular == True], color="black")
    b, a = np.polyfit(dataset[dataset.is_hyper_popular == True][metric], dataset[dataset.is_hyper_popular == True].streams, deg=1)
    ax[i][1].plot(dataset[dataset.is_hyper_popular == True][metric], a + b * dataset[dataset.is_hyper_popular == True][metric], color="purple")
    ax[i][1].set_xticks(range(0, 101, 10))
    ax[i][1].set_title(f"{metric.title()} vs Number of Streams for Hyper-Popular Songs", color="white")
    ax[i][1].set_ylabel('Number of Streams (Billions)', color='white')
    ax[i][1].set_xlabel(f'{metric.title()} (%)', color='white')
    ax[i][1].grid(visible = True)
    ax[i][1].set_axisbelow(True)
    ax[i][1].tick_params(axis='x', colors='white')
    ax[i][1].tick_params(axis='y', colors='white')

plt.tight_layout()
plt.show()
```

When analyzing each of the visualizations, we can see there is significant scattering of data points across each of its applicable range of values. This is confirmed through the large amount of vertical variance for the number of streams across different percentage point ratings (for each spotify metric) which makes it very difficult to assume any kind of linear relationship. Nonetheless, we have still plotted lines of best fit to help illustrate the relatively flat slopes that depict this lack of correlation. A noticeable differences between the visualizations for popular and hyper-popular songs are in the sample size (number of songs) which contribute to slightly exaggerated slopes due to some of the data points having greater leverage over the line of best fit. Adding onto this, we can see that there is an absence of songs across different sections of the scatter-plots in the hyper-popular domain, such as in speechiness and liveness where there aren't any songs reporting with values above ~45% and ~50. This may suggest that it may be rarer on average for songs to become hyper-popular if they contain exclusively speech-like content or are recorded with a live audience. This notion is most emphasized when observing the scatter plot of instrumentalness for hyper-popular songs, where the maximum value recorded for this metric was 10%. The significance behind this is that there were no hyper-popular songs that were strictly or predominantly instrumental, showing that they all had vocal content.

In our final visualization, we are able to discern a relatively apparent linear relationship between the number of streams and the number of playlist adds. There is a positive correlation between the two, suggesting that as songs become increasingly popular around the globe, so does their presence on playlists worldwide. This intuitively makes sense and confirms our underlying assumption about the relationship between these two variables of interest. What we found noteworthy was how the linear relationship weakens as the number of streams increases, as data points are more tightly clustered from 0 to 0.5 billion streams, and become progressively more dispersed from there onwards. It would be most reliable to predict the number of playlist adds with respect to the number of streams only where the linear correlation is strongest. Interestingly, the rate of growth for playlists adds has slowed down in the hyper-popular domain as compared to the popular domain, signifying that as the number of streams for a song approaches a certain range of values, that there would be a plateau in the number of playlist adds.

In conclusion, our project focused on analyzing top-performing songs on Spotify's platform in 2023, aiming to uncover valuable insights for both artists and Spotify. Through our exploration of guiding questions, we discovered several key findings.

Second, our analysis of Spotify-specific metrics and their correlation with the number of streams revealed inconclusive evidence of strong linear relationships, suggesting that popularity on Spotify is influenced by various factors, making it challenging to predict success based solely on these metrics.

Lastly, we observed a positive correlation between the number of streams and the number of playlist adds, indicating that as songs become more popular, they tend to appear on more playlists. However, this relationship weakens as the number of streams increases, and playlist adds start to taper off for hyper-popular songs.

In the future, research in this field could explore the influence of other external factors, such as marketing strategies, artist popularity, or cultural events, on the streaming performance of songs.
